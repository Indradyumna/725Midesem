```python
import torch
import numpy as np
import matplotlib.pyplot as plt
xlist = np.arange(-10,10,0.1)
ylist = []
for x in xlist:
    X = torch.FloatTensor([x])
    W1 = torch.FloatTensor([[1,1,-1,-1]])
    B1 = torch.FloatTensor([0,-1,-1,0])
    relu = torch.nn.ReLU()
    W2 = torch.FloatTensor([1,-1,1,-1])
    B2 = torch.FloatTensor([0])
    out = relu(X@W1+B1)@W2
    ylist.append(out)
plt.plot(xlist,ylist) 
```
